```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

airq_data = pd.read_csv('world_air_quality.csv', delimiter=';') #read data and organize into columns using delimiter
airq_data
```

```python
#remove spaces from left and right of strings
airq_data.columns = airq_data.columns.str.lstrip()
airq_data.columns = airq_data.columns.str.rstrip()
airq_data
```

```python
# drop unnecessary columns
airq_data = airq_data.drop(['Country Code','City','Location','Coordinates','Source Name', 'Last Updated'],axis=1)
airq_data
```

```python
#rename "Country Label" to 'Country'
airq_data.rename(columns={'Country Label': 'Country'}, inplace=True)
airq_data
```

```python
# Drop rows with NaN values in 'Country', 'Pollutant', or 'Value' columns
airq_data.dropna(subset=['Country', 'Pollutant', 'Value'], inplace=True)

print(airq_data)
```

```python
# Update the DataFrame to keep only rows where the "Unit" is either "ppm" or "ug/cm^3"
airq_data = airq_data[(airq_data['Unit'] == 'ppm') | (airq_data['Unit'] == 'µg/m³')]
#airq_data

# Drop rows with negative values in the 'Value' column
airq_data = airq_data[airq_data['Value'] >= 0]

airq_data
```

```python
# Convert all data into one unit for uniformity

# Molecular weights of gases (in g/mol)
molecular_weights = {
    'CO': 28.01,     # Carbon Monoxide
    'NO': 30.01,     # Nitric Oxide
    'NO2': 46.01,    # Nitrogen Dioxide
    'NOX': 46.01,    # Nitrogen Oxides (Sum of NO and NO2)
    'O3': 48.00,     # Ozone
    'PM2.5': 12.01,  # Particulate Matter ≤ 2.5 µm
    'SO2': 64.07     # Sulfur Dioxide
}

# Identify pollutants measured in ppm and convert their values to µg/m³
for index, row in airq_data.iterrows():
    pollutant = row['Pollutant']
    if pollutant in molecular_weights and row['Unit'] == 'ppm':
        molecular_weight = molecular_weights[pollutant]
        # Convert ppm to µg/m³ using the conversion factor: 1 ppm = 1 µg/m³ (assuming 1 m³ volume)
        conversion_factor = molecular_weight * 1e6
        airq_data.at[index, 'Value'] *= conversion_factor

# Drop the 'Unit' column
airq_data = airq_data.drop(['Unit'],axis=1)

airq_data
```

```python
# Data Discretization

# Quartiles for the 'Value' column
Q1 = airq_data['Value'].quantile(0.25)
Q3 = airq_data['Value'].quantile(0.75)

# quartile boundaries
quarts = [0.0, Q1, Q3, float('inf')]

# Labels for quartile categories
labels = ['Low', 'Medium', 'High']

# New column 'Pollution Category' based on quartile categories
airq_data['Pollution Category'] = pd.cut(airq_data['Value'], bins=quarts, labels=labels, include_lowest=True)

# Drop the 'Value' column
airq_data = airq_data.drop(['Value'],axis=1)

airq_data
```

```python
from sklearn.preprocessing import OneHotEncoder
import pycountry_convert as pc

# Function to map country codes to continents
def map_country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        continent_name = pc.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except KeyError:
        return None

# Mapping countries to continents
airq_data['Continent'] = airq_data['Country'].apply(map_country_to_continent)

# categorical columns for one-hot encoding
categorical_cols = ['Continent', 'Pollutant', 'Pollution Category']

encoder = OneHotEncoder(sparse_output=False)
# sparse_output=False ensures the output is a dense array

# Fit and transform
onehot_encoded_data = encoder.fit_transform(airq_data[categorical_cols])

# Converting the encoded data into a DataFrame with proper column names
encoded_df = pd.DataFrame(onehot_encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

encoded_df = encoded_df.dropna()

# encoded data with headings
print(encoded_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder

# Concatenating multiple pollution columns into a single target variable
pollution_categories = [col for col in encoded_df.columns if col.startswith('Pollution Category_')]

y = encoded_df[pollution_categories].apply(lambda x: '_'.join(x.astype(str)), axis=1)

# Converting y to single-label format
label_encoder = LabelEncoder()
y_single_label = label_encoder.fit_transform(y)

# Features (independent variables)
continent_columns = [col for col in encoded_df.columns if col.startswith('Continent_')]
pollutants = [col for col in encoded_df.columns if col.startswith('Pollutant_')]
X = encoded_df[continent_columns + pollutants]

# range of k values to evaluate
k_values = range(5, 25)

# Performing stratified K-fold cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Creating empty lists to store k values and accuracy scores
k_list = []
accuracy_list = []

# Iterating over k values
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    accuracy = cross_val_score(knn, X, y_single_label, cv=cv, scoring='accuracy').mean()
    k_list.append(k)
    accuracy_list.append(accuracy)

# Creating a DataFrame from the lists
accuracy_df = pd.DataFrame({'k': k_list, 'Accuracy': accuracy_list})

# Printing the accuracy values in a table
print("Accuracy values for all k values:")
print(accuracy_df)

# Plotting the k values vs. accuracy scores
plt.figure(figsize=(8, 6))
plt.plot(k_values, accuracy_list, marker='o', label='Accuracy')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy Score')
plt.title('KNN Accuracy vs. Number of Neighbors (k)')
plt.xticks(k_values)  # Setting x-axis ticks for all k values
plt.grid(True)
plt.legend()
plt.show()
```
